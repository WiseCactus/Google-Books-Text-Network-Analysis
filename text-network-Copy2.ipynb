```python
from nltk.stem import SnowballStemmer
from sklearn.feature_extraction import text
import networkx as nx
import re
from collections import Counter
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
import nltk
from nltk import pos_tag, word_tokenize


# Gets rid of stopwords and symbols.
from textblob import Word
from nltk.corpus import wordnet
from PyDictionary import PyDictionary
from hunspell import Hunspell

dictionary=PyDictionary()

stop_words = stopwords.words('english')

stopwords_dict = Counter(stop_words)


# We will use these functions for text preprocessing.

def pos_tagging(text):
    return pos_tag(word_tokenize(text))[0][1]

def check_stopword(corpus):
    if corpus in stop_words:
        return ""
    return corpus

def stemmed(corpus):
    h = Hunspell()
    n=h.stem(corpus)
    if len(n)>1:
            return n[1]
    if (len(n)>0):
        return n[0]
    else:
        return corpus
        #return " ".join(stemmer.stem(word) for word in corpus.split(" "))


# This function will remove all symbols from the text.

def remove_symbols(corpus):
        corpus = re.sub(r'\W+', ' ',corpus)
        corpus = re.sub(" \d+", " ", corpus)
        corpus = re.sub(' s ', ' ',corpus)
        corpus = re.sub(' ve ', ' ',corpus)
        corpus = re.sub(' m ', ' ',corpus)
        corpus = re.sub(' t ', ' ',corpus)
        return corpus
# This function will text      

def ifValid(corpus):
    
    if corpus is None:
        return "" 
    
    h = Hunspell()
    if h.spell(corpus) and len(corpus)>1:
        return corpus
    if h.spell(corpus.lower()) and len(corpus)>1:
        return corpus
    else:
        return ""
        
        
   
def correct(corpus):
    if (dictionary.meaning(corpus)) is None:
        for i in range(len(corpus)-1):
            word = Word(corpus[i:])
            spelling = (word.spellcheck()[0][0])
            if (spelling != corpus[i:]):
                return spelling
    else:
        return corpus
        
```
